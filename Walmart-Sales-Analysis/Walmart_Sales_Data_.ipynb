```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
basket = df.groupby(['Customer_ID', 'Product_Name'])['Purchase_Amount'].count().unstack().fillna(0)

# Convert values into 0 (not purchased) or 1 (purchased)
basket = basket.applymap(lambda x: 1 if x > 0 else 0)

```

```python
# Find frequent itemsets with a minimum support of 2% (0.02)
frequent_itemsets = apriori(basket, min_support=0.02, use_colnames=True)

# Generate Association Rules (Min confidence = 30%)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)

# Display the top rules
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].sort_values(by="lift", ascending=False).head(10)
```
